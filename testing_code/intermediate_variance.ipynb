```python
import numpy as np
import matplotlib.pyplot as plt


def get_average_standard_deviation_across_features(a):

    # number of predictions, number of test samples, number of features
    # a = np.random.rand(2, 10, 3)
    b = np.mean(a, axis=0)
    c = a - b
    d = np.sqrt(np.mean(c**2, axis=0))
    e = np.mean(d, axis=1)
    return e

# plot to test the function
v = np.random.rand(2, 3, 2)

sd = get_standard_deviation(v)

mean = np.mean(v, axis=0)

# plot the points and the mean

plt.scatter(v[0, :, 0], v[0, :, 1], color='blue')

plt.scatter(v[1, :, 0], v[1, :, 1], color='red')

plt.scatter(mean[:, 0], mean[:, 1], color='green', s=sd*100)














```
